```python
#Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#loading Complete Dataset
pd.set_option('display.max_columns', 500)
full_df=pd.read_csv("FullData.csv")
full_df.head()
```

```python
#Creating new dataset with the coumns we'll be using
df=full_df.loc[:,"Weak_foot":"GK_Reflexes"]
```

```python
len(full_df.Club_Position.unique())
```

```python
df.describe()
```

```python
#Importing Kmeans for clustering & fitting to df
from sklearn.cluster import KMeans
kmeans_model= KMeans(n_clusters=4,random_state=1)
kmeans_model.fit(df)
labels=kmeans_model.labels_
labels
```

```python
# Imorting and using PCA 

from sklearn.decomposition import PCA
pca=PCA(2)
pca_df=pca.fit_transform(df)
```

```python
pca_df
```

```python
# Plotting results
col=sns.color_palette("muted",n_colors=4)
sns.set_style("white")
sns.scatterplot(x=pca_df[:,0],y=pca_df[:,1],hue=labels,palette=col)
```

Now I'll be choosing some players from different positions to study the segregation done by kmeans clustering

```python
#Forwards

Cristiano_Ronaldo=df.loc[full_df["Name"]=="Cristiano Ronaldo"].values.tolist() #LW (Left Winger)

Lionel_Messi=df.loc[full_df["Name"]=="Lionel Messi"].values.tolist()       #RW (Right Winger)
```

```python
#Mid-Feilders

Mesut_Özil = df.loc[full_df["Name"]=="Mesut Özil"].values.tolist() #CAM (Center Attacking Midfeilder)

Sergio_Busquets =df.loc[full_df["Name"]=="Sergio Busquets"].values.tolist() #CDM (Center Defending Midfeilder)
```

```python
#GoalKeepers

Manuel_Neuer=df.loc[full_df["Name"]=="Manuel Neuer"].values.tolist() #Goal Keeper

De_Gea=df.loc[full_df["Name"]=="De Gea"].values.tolist()       #Goal Keeper
```

```python
#Backwards

Toby_Alderweireld = df.loc[full_df["Name"]=="Toby Alderweireld"].values.tolist()  #CB (Center Backward)

David_Luiz = df.loc[full_df["Name"]=="David Luiz"].values.tolist()        #CB (Center Backward)
```

```python
CR=kmeans_model.predict(Cristiano_Ronaldo)
LM=kmeans_model.predict(Lionel_Messi)
MO=kmeans_model.predict(Mesut_Özil)
SB=kmeans_model.predict(Sergio_Busquets)
TA=kmeans_model.predict(Toby_Alderweireld)
DL=kmeans_model.predict(David_Luiz)
MN=kmeans_model.predict(Manuel_Neuer)
DG=kmeans_model.predict(De_Gea)

print(f"\n Cristiano Ronaldo belongs to label {CR} & Lionel Messi belongs to label {LM} : Forwards")
print(f"\n Mesut Özil belongs to label {MO} & Sergio Busquets belongs to label {SB} : Mid-Feilders")
print(f"\n Toby Alderweireld belongs to label {TA} & David Luiz belongs to label {DL} : Backwards")
print(f"\n Manuel Neuer belongs to label {MN} De Gea belongs to label {DG} : Goal Keepers ")

```

**You must be thinking that when all the Forward players,Backward players & Goal keepers are falling under same labels then why are the Mid players falling under different category** ,

This is because there are two types of mid players :

1. Mid-forwards : They have the skill set to play mid as well as forwards; CAM (Center Attacking Midfeilder)
2. Mid-backwards : They have the skill set to play mid as well as backwards ; CDM (Center Defending Midfeilder)

That's the reason why :

* Mesut Özil (CAM) falls under the label :1 which consists all the forward players  
* Sergio Busquets (CDM) falls under the label :2 which consists all the backward players
